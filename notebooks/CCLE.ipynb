### "Melt" original colnames into a single column as 'cell_id', and original cloumn values as single column 'expression'

```r
ccle <- melt(ccle, id.vars = "Name", variable.name = "cell_id", value.name = "expression", variable.factor = F, value.factor = F)
head(ccle)
```

### Create 'cell_id2' column from  'cell_id' prefix

```r
ccle[, cell_id2 := tstrsplit(cell_id, "_", fixed = T, keep = 1)]
head(ccle)
```

### Create 'tissue' column from 'cell_id suffix

```r
ccle[cell_id2 != cell_id, tissue := substr(cell_id, nchar(cell_id2) + 2, nchar(cell_id))]
head(ccle)
```

### Remove 'cell_id' column and rename 'cell_id2' as 'cell_id'

### Create 'ENSG' column with version suffix stripped off 'Name'

```r
ccle[, ENSG := tstrsplit(Name, ".", fixed = T, keep = 1)]
head(ccle)
```

### Remove 'Name' column

```r
ccle[, Name := NULL]
head(ccle)
```

## Now connect to TCRD to get protein_ids and corresponding ENSGs

```r
library(RMySQL, quietly = T)
DBHOST <- 'localhost'
DBNAME <- 'tcrd6'
DBUSER <- 'smathias'

dbconn <- dbConnect(MySQL(), host=DBHOST, dbname=DBNAME, user=DBUSER)
```

```r
sql <- "SELECT p.id AS protein_id, x.value AS ensg 
        FROM protein p, xref x 
        WHERE p.id = x.protein_id AND x.xtype = 'ENSG'"
prots <- dbGetQuery(dbconn, sql)
dbDisconnect(dbconn)
rm(dbconn)
setDT(prots) # convert data.frame to data.table
```

```r
head(prots)
```

```r
nrow(prots)
```

### Join ccle to prots on ENSGs

```r
dt <- merge(ccle, prots, by.x = "ENSG", by.y = "ensg", all.x = T, allow.cartesian = T)
```

```r
head(dt)
```

```r
nrow(dt)
```

### Remove rows with no protein_id

```r
dt <- dt[!is.na(protein_id)]
nrow(dt)
```

### Remove 'ENSG' column

```r
dt[, ENSG := NULL]
head(dt)
```

```r
setcolorder(dt, c("protein_id", "cell_id", "tissue", "expression"))
dt <- unique(dt)
head(dt)
```

```r
nrow(dt)
```

### Write data table to TSV

```r
OUTPUT_FILE <- '/Users/smathias/TCRD/data/CCLE/CCLE.tsv'
fwrite(dt, file = OUTPUT_FILE, quote = T, sep = "\t", col.names = T, row.names = F, na = "None")
```

```r
GZIP_FILE <- '/Users/smathias/TCRD/data/CCLE/CCLE.tsv.gz'
if(file.exists(GZIP_FILE)) {
  file.remove(GZIP_FILE)
}
system(sprintf("gzip -9v %s", OUTPUT_FILE))
```
